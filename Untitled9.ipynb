```python
def display_genome_1kb(genome_sequence):
  """
  Presents a genome sequence in blocks of 1 kilobase (1000 bases)
  with line numbering for readability.

  Args:
    genome_sequence (str): The DNA sequence to display.
  """

  for i in range(0, len(genome_sequence), 1000):
    chunk = genome_sequence[i : i + 1000]
    line_number = i // 1000 + 1
    print(f"Line {line_number}: {chunk}")

# Example usage (replace with your actual genome sequence)
example_genome = "ATCG" * 250  # Creates a 1000 base pair sequence
display_genome_1kb(example_genome)
```

```python
def nucleotide_statistics(genome_sequence):
  """
  Calculates and prints the count and percentage of each nucleotide (A, T, G, U)
  in a given genome sequence.

  Args:
    genome_sequence (str): The DNA/RNA sequence to analyze.
  """

  total_bases = len(genome_sequence)

  # Count each nucleotide
  count_A = genome_sequence.count('A')
  count_T = genome_sequence.count('T')
  count_G = genome_sequence.count('G')
  count_U = genome_sequence.count('U')

  # Calculate percentages
  percent_A = (count_A / total_bases) * 100
  percent_T = (count_T / total_bases) * 100
  percent_G = (count_G / total_bases) * 100
  percent_U = (count_U / total_bases) * 100

  # Print the results
  print("Nucleotide Statistics:")
  print(f"A: Count = {count_A}, Percentage = {percent_A:.2f}%")
  print(f"T: Count = {count_T}, Percentage = {percent_T:.2f}%")
  print(f"G: Count = {count_G}, Percentage = {percent_G:.2f}%")
  print(f"U: Count = {count_U}, Percentage = {percent_U:.2f}%")

# Example usage (replace with your actual genome sequence)
example_genome = "ATCG" * 250
nucleotide_statistics(example_genome)
```

```python
from Bio import Entrez
from Bio import SeqIO

# Replace with your email
Entrez.email = "ltashev@abv.com"

# Fetch the M. genitalium genome from NCBI
handle = Entrez.efetch(db="nucleotide", id="NC_000908.1", rettype="fasta", retmode="text")
record = SeqIO.read(handle, "fasta")
handle.close()

# Store the DNA sequence as a string
mg_dna = str(record.seq)

def calculate_gc_content(dna_sequence):
  """
  Calculates the GC content of a DNA sequence.

  Args:
    dna_sequence (str): The DNA sequence to analyze.

  Returns:
    float: The GC content as a percentage.
  """

  total_bases = len(dna_sequence)
  count_G = dna_sequence.count('G')
  count_C = dna_sequence.count('C')

  gc_content = ((count_G + count_C) / total_bases) * 100
  return gc_content

# Calculate and print the GC content of M. genitalium
gc_content = calculate_gc_content(mg_dna)
print(f"The GC content of M. genitalium is: {gc_content:.2f}%")
```

```python
from Bio import Entrez
from Bio import SeqIO

# Replace with your email
Entrez.email = "ltashev@abv.com"

# Fetch the M. genitalium genome from NCBI
handle = Entrez.efetch(db="nucleotide", id="NC_000908.1", rettype="fasta", retmode="text")
record = SeqIO.read(handle, "fasta")
handle.close()

# Store the DNA sequence as a string
mg_dna = str(record.seq)

def find_motifs(dna_sequence, motif):
  """
  Finds all occurrences of a specific motif in a DNA sequence.

  Args:
    dna_sequence (str): The DNA sequence to search.
    motif (str): The motif to find.

  Returns:
    list: A list of positions (start indices) where the motif is found.
  """

  positions = []
  index = 0
  while index < len(dna_sequence):
    index = dna_sequence.find(motif, index)
    if index == -1:
      break
    positions.append(index)
    index += 1  # Move to the next position after the found motif

  return positions

# Example usage: Search for the motif "ATGATC"
motif_to_find = "ATGATC"
motif_positions = find_motifs(mg_dna, motif_to_find)

if motif_positions:
  print(f"The motif '{motif_to_find}' was found at the following positions:")
  for position in motif_positions:
    print(position)
else:
  print(f"The motif '{motif_to_find}' was not found in the M. genitalium genome.")
```

```python
#!pip install rdkit-pypi

import py3Dmol
from rdkit import Chem
from rdkit.Chem import AllChem

# Generate Lysine molecule
mol = Chem.MolFromSmiles('C(CCN)CC(C(=O)O)N')
AllChem.EmbedMolecule(mol)

# Convert to PDB format
pdb_block = Chem.MolToPDBBlock(mol)

# Visualize using Py3Dmol
view = py3Dmol.view()
view.addModel(pdb_block, 'pdb')
view.setStyle({'stick': {}})
view.zoomTo()
view.show()
```
