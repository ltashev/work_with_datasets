```python
import random

# Number of random points to generate
num_points = 100000

# Initialize count of points inside the circle
inside_circle = 0

# Generate random points and check if they are inside the unit circle
for _ in range(num_points):
  x = random.random()
  y = random.random()
  if x**2 + y**2 <= 1:
    inside_circle += 1

# Estimate Pi
pi_estimate = 4 * inside_circle / num_points

print("Estimated value of Pi:", pi_estimate)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяне на факторите и нивата
concentration_levels = [0.1, 0.2, 0.3]  # Моларност на прекурсора (M)
temperature_levels = [60, 80, 100]  # Температура на реакцията (°C)

# Функция за симулиране на размера на наночастиците
def synthesize_zno(concentration, temperature):
    # Примерна функция за симулация на размера на наночастиците
    size = 100 / (concentration * temperature)  # Примерна зависимост
    return size

# Събиране на данни
data = []
for conc in concentration_levels:
    for temp in temperature_levels:
        size = synthesize_zno(conc, temp)
        data.append((conc, temp, size))

# Преобразуване на данните в numpy масив за по-лесна обработка
data = np.array(data)

# Визуализация на резултатите
fig, ax = plt.subplots()
for conc in concentration_levels:
    sizes = data[data[:, 0] == conc][:, 2]
    ax.plot(temperature_levels, sizes, label=f'Concentration {conc} M')

ax.set_xlabel('Temperature (°C)')
ax.set_ylabel('Particle Size (nm)')
ax.set_title('Effect of Concentration and Temperature on ZnO Nanoparticle Size')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяне на параметрите
nanoparticle_concentration = [0.1, 0.2, 0.3]  # Концентрация на наночастиците (%)
mixing_time = [10, 20, 30]  # Време за смесване (минути)

# Функция за симулиране на смесването
def mix_nanoparticles_in_chocolate(concentration, time):
    # Примерна функция за симулация на хомогенността на смесването
    homogeneity = 100 - (concentration * time)  # Примерна зависимост
    return homogeneity

# Събиране на данни
data = []
for conc in nanoparticle_concentration:
    for time in mixing_time:
        homogeneity = mix_nanoparticles_in_chocolate(conc, time)
        data.append((conc, time, homogeneity))

# Преобразуване на данните в numpy масив за по-лесна обработка
data = np.array(data)

# Визуализация на резултатите
fig, ax = plt.subplots()
for conc in nanoparticle_concentration:
    homogeneities = data[data[:, 0] == conc][:, 2]
    ax.plot(mixing_time, homogeneities, label=f'Concentration {conc} %')

ax.set_xlabel('Mixing Time (minutes)')
ax.set_ylabel('Homogeneity (%)')
ax.set_title('Effect of Nanoparticle Concentration and Mixing Time on Chocolate Homogeneity')
ax.legend()
plt.show()
```
