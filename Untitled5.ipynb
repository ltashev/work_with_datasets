```python
import pandas as pd
import random

# Генериране на случайни данни
num_samples = 100  # Брой на примерните записи
data = {
    'stress_level': [random.randint(1, 10) for _ in range(num_samples)],
    'fatigue_level': [random.randint(1, 5) for _ in range(num_samples)],
    'sleep_quality': [random.uniform(4, 9) for _ in range(num_samples)],
    'happiness_score': [random.uniform(0, 1) for _ in range(num_samples)],
    'steps_count': [random.randint(1000, 15000) for _ in range(num_samples)]
}

# Създаване на DataFrame
df = pd.DataFrame(data)

# Запис на данните в CSV файл
df.to_csv('psychophysiological_dataset.csv', index=False)
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Създаване на хистограма
plt.hist(df['stress_level'], bins=10, edgecolor='black')
plt.xlabel('Стрес ниво')
plt.ylabel('Брой наблюдения')
plt.title('Разпределение на стрес нивото')
plt.show()
```

```python
import seaborn as sns

# Създаване на box plot
sns.boxplot(x='sleep_quality', data=df)
plt.xlabel('Качество на съня')
plt.title('Box plot на качеството на съня')
plt.show()
```
